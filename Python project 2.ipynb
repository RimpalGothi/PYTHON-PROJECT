# 1. Find the total number of crimes recorded in the CRIME table.

```python
%%sql

SELECT COUNT(*) as total_crimes
FROM CHICAGO_CRIME_DATA;
```

# 2. List community areas with per capita income less than 11000.

```python
%%sql
SELECT COMMUNITY_AREA_NAME, PER_CAPITA_INCOME 
FROM CENSUS_DATA
WHERE PER_CAPITA_INCOME < 11000;
```

# 3. List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis)

```python
%%sql

SELECT CASE_NUMBER
FROM CHICAGO_CRIME_DATA
WHERE PRIMARY_TYPE = 'OFFENSE INVOLVING CHILDREN'
```

# 4. List all kidnapping crimes involving a child?¶

```python
%%sql

SELECT DISTINCT PRIMARY_TYPE
FROM CHICAGO_CRIME_DATA
WHERE PRIMARY_TYPE = 'KIDNAPPING' AND DESCRIPTION LIKE '%CHILD%'
```

# 5. What kinds of crimes were recorded at schools?

```python
%%sql

SELECT DISTINCT PRIMARY_TYPE
FROM CHICAGO_CRIME_DATA
WHERE Location_DESCRIPTION LIKE '%School%'
```

# 6. List the average safety score for each type of school.

```python
%%sql

SELECT NAME_OF_SCHOOL, AVG(SAFETY_SCORE) AS AVERAGE_SAFETY_SCORE
FROM CHICAGO_PUBLIC_SCHOOLS
GROUP BY NAME_OF_SCHOOL;
```

# 7. List 5 community areas with highest % of households below poverty line

```python
%%sql
SELECT COMMUNITY_AREA_NAME, PERCENT_HOUSEHOLDS_BELOW_POVERTY
FROM CENSUS_DATA
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC
LIMIT 5;
```

# 8. Which community area is most crime prone?

```python
%%sql
SELECT COMMUNITY_AREA_NUMBER, COUNT(*) AS CRIME_COUNT
FROM CHICAGO_CRIME_DATA
GROUP BY COMMUNITY_AREA_NUMBER
ORDER BY CRIME_COUNT DESC
LIMIT 1;
```

# 9. Use a sub-query to find the name of the community area with highest hardship index

```python
%%sql

SELECT COMMUNITY_AREA_NAME
FROM CENSUS_DATA
WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM CENSUS_DATA);
```

# 10. Use a sub-query to determine the Community Area Name with most number of crimes?

```python
%%sql
SELECT DISTINCT LOCATION_DESCRIPTION
FROM CHICAGO_CRIME_DATA
WHERE LOCATION_DESCRIPTION = (SELECT LOCATION_DESCRIPTION
                       FROM CHICAGO_CRIME_DATA
                       GROUP BY LOCATION_DESCRIPTION
                       ORDER BY COUNT(*) DESC
                       LIMIT 1);
```

```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv (url)
df.head()
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(url, names = headers)
df.head()
```

```python
import numpy as np
```

```python
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
df.head(2)
```

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
df.head(2)
```

```python
avg_stroke = df["stroke"].astype("float").mean(axis=0)
print("Average of stroke:", avg_stroke)
```

```python
df["stroke"].replace(np.nan, avg_stroke, inplace=True)
df.head()
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
df.head()
```

```python
df['city-L/100km'] = 235/df["city-mpg"]
df.head()
```

```python
df.rename(columns={"highway-mpg": "highway-L/100km"}, inplace=True)
df.head(2)
```

```python
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df['height'] = df['height']/df['height'].max()
df.head(2)
```

```python
df["horsepower"]=df["horsepower"].astype(int, copy=True)
```
